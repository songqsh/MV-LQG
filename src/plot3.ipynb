# Plot value for a partially observable system

```python
from mpl_toolkits.mplot3d import Axes3D
import matplotlib.pyplot as plt
from matplotlib import cm
import numpy as np
```

```python
#parameters
x = 1
T = 2
s = 1
```

```python
v_1 = lambda sg, eta: (np.power(x,2) + np.power(eta, 2))/(1+T-s) + \
    np.power(sg,2)*np.log(1+T-s) + (1-np.power(eta,2))*s+ (1 - np.power(sg,2))*T
```

```python
v_1(.2,.4)
```

```python
sg = np.linspace(0,1,11)
eta = np.linspace(0,1,11)
sg, eta = np.meshgrid(sg,eta)

value = v_1(sg, eta)
```

```python
fig = plt.figure()
ax = fig.gca(projection='3d')

# Plot the surface.
surf = ax.plot_surface(sg, eta, value, cmap=cm.coolwarm)

ax.set_xlabel('$\hat\sigma$')
ax.set_xlim(0, 1)
ax.set_ylabel('$\hat\eta$')
ax.set_ylim(0, 1)
ax.set_zlabel('$V^*$')

# Add a color bar which maps values to colors.
fig.colorbar(surf, shrink=0.5, aspect=5)
#plt.show()
plt.savefig('plot3.png')
```
